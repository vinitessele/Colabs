🔠 Pipeline 1: Texto / Geração / Tradução

```python
!pip install -q transformers torch sentencepiece

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_id = "rubuntu/gemma-2-9b-it-SimPO-Jopara-V3.4"

tokenizer_text = AutoTokenizer.from_pretrained(model_id)
model_text = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

pipe_text = pipeline(
    "text-generation",
    model=model_text,
    tokenizer=tokenizer_text,
    return_full_text=False
)

prompts = [
    "Traduza para o guarani: A escola está aberta.",
    "Traduza para o guarani: Vamos à floresta.",
    "Traduza para o guarani: Eu te amo.",
    "Traduza para o guarani: Onde você mora?",
    "Traduza para o guarani: Bom dia, como você está?",
    "Traduza para o guarani: Estamos em paz.",
    "Explique em Guarani: amizade é confiança",
    "Explique em Guarani: educação é o caminho para o futuro.",
    "Explique em Guarani: precisamos preservar a floresta.",
    "Explique em Guarani: respeito aos mais velhos é sagrado.",
    "Explique em Guarani: a língua Guarani é parte de nossa identidade.",
    "Explique em Guarani: água é vida.",
    "Explique em Guarani: saúde vem da natureza.",
    "Explique em Guarani: aprender é crescer.",
    "Traduza para o guarani: A criança está brincando.",
    "Traduza para o guarani: Minha casa é pequena, mas feliz.",
    "Traduza para o guarani: O sol está brilhando.",
    "Traduza para o guarani: Eu venho da aldeia.",
    "Traduza para o guarani: Vamos aprender juntos.",
    "Traduza para o guarani: A natureza é sagrada."
]

for prompt in prompts:
    res = pipe_text(prompt, max_new_tokens=64, temperature=0.7, top_p=0.9)
    print(f"🗣️ Prompt: {prompt}\n➡️ Resposta: {res[0]['generated_text']}\n")
```

🎙️ Pipeline 2: Áudio → Texto (ASR)

```python
!pip install -q transformers torchaudio

from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import torch, torchaudio

model_asr = "ivangtorre/wav2vec2-xlsr-300m-guarani"

processor = Wav2Vec2Processor.from_pretrained(model_asr)
model_asr = Wav2Vec2ForCTC.from_pretrained(model_asr)

def transcrever(audio_path):
    waveform, sr = torchaudio.load(audio_path)
    input_values = processor(waveform.squeeze(), sampling_rate=sr, return_tensors="pt").input_values
    with torch.no_grad():
        logits = model_asr(input_values).logits
    pred_ids = torch.argmax(logits, dim=-1)
    return processor.batch_decode(pred_ids)[0]

# Exemplo: substitua "audio1.wav" e "audio2.wav" pelos seus arquivos de fala em Guarani.
for audio in ["audio1.wav", "audio2.wav"]:
    print(f"🎧 Transcrição ({audio}): {transcrever(audio)}")
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch
import pandas as pd
from difflib import SequenceMatcher

# Modelos
model_a_path = "Emilio407/guarani-jopara-gemma-2-2b-it-v1"  # seu fine-tuned
model_b_path = "rubuntu/gemma-2-9b-it-SimPO-Jopara-V3.4"    # modelo base maior

# Carrega os pipelines
pipe_a = pipeline("text-generation",
                  model=AutoModelForCausalLM.from_pretrained(model_a_path),
                  tokenizer=AutoTokenizer.from_pretrained(model_a_path),
                  return_full_text=False)

pipe_b = pipeline("text-generation",
                  model=AutoModelForCausalLM.from_pretrained(model_b_path, torch_dtype=torch.float16, device_map="auto"),
                  tokenizer=AutoTokenizer.from_pretrained(model_b_path),
                  return_full_text=False)

# Prompts originais para usar no ensemble
prompts = [
    "Traduza para o guarani: A escola está aberta.",
    "Traduza para o guarani: Vamos à floresta.",
    "Traduza para o guarani: Eu te amo.",
    "Explique em Guarani: amizade é confiança",
    "Explique em Guarani: saúde vem da natureza.",
    "Traduza para o guarani: O sol está brilhando.",
]

# Função simples de heurística (exemplo: menor distância entre entrada e resposta → rejeitar eco)
def score_response(prompt, response):
    ratio = SequenceMatcher(None, prompt.lower(), response.lower()).ratio()
    return 1 - ratio  # quanto menor a similaridade com o prompt, melhor

# Geração e seleção
data = []

for prompt in prompts:
    res_a = pipe_a(prompt, max_new_tokens=64, temperature=0.7, top_p=0.9)[0]['generated_text']
    res_b = pipe_b(prompt, max_new_tokens=64, temperature=0.7, top_p=0.9)[0]['generated_text']

    score_a = score_response(prompt, res_a)
    score_b = score_response(prompt, res_b)

    best = res_a if score_a > score_b else res_b

    data.append({
        "instruction": prompt.split(":")[0] + ":",
        "input": prompt.split(":")[1].strip(),
        "output": best.strip()
    })

    print(f"\n🗣️ {prompt}")
    print(f"🔹 A: {res_a.strip()} (score: {score_a:.2f})")
    print(f"🔸 B: {res_b.strip()} (score: {score_b:.2f})")
    print(f"✅ Escolhido: {best.strip()}")
```
