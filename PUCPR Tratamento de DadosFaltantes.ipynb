```python
# Importar bibliotecas
import pandas as pd
from google.colab import files

# Passo 1: Fazer upload do arquivo
print("üìÅ Fa√ßa upload do arquivo chamado 'clientes2.csv':")
uploaded = files.upload()

# Passo 2: Ler os dados do arquivo
df = pd.read_csv("clientes2.csv")

# Passo 3: Visualiza√ß√£o geral
print("‚úÖ Primeiras linhas do arquivo:")
display(df.head())

print("\nüìä Informa√ß√µes do DataFrame:")
print(df.info())

print("\nüîç Contagem de valores ausentes por coluna:")
print(df.isnull().sum())
```

```python
# Preencher valores ausentes em 'idade' com a m√©dia
df['idade'] = df['idade'].fillna(df['idade'].mean())

# Preencher sexo com moda
df['sexo'] = df['sexo'].fillna(df['sexo'].mode()[0])

# Visualizar ap√≥s tratamento
print("‚úÖ Dados ap√≥s preenchimento:")
display(df)
```

```python
import pandas as pd
from google.colab import files

# Upload do arquivo
print("üìÅ Fa√ßa upload do arquivo 'clima_julho.csv'")
uploaded = files.upload()

# Leitura do CSV
df = pd.read_csv('clima_julho.csv')

# Visualiza√ß√£o inicial
print("üìä Dados brutos:")
display(df)

# Contar valores ausentes
print("üîç Valores ausentes por coluna:")
print(df.isnull().sum())

# Preenchimento com m√©todos diferentes
df_temp = df.copy()
df_temp['temperatura'] = df_temp['temperatura'].interpolate()
df_temp['chuva_mm'] = df_temp['chuva_mm'].fillna(0)  # assuma que faltantes = sem chuva

# Mostrar resultado
print("‚úÖ Ap√≥s preenchimento (temp = interpola√ß√£o, chuva = 0):")
display(df_temp)

# Salvar CSV limpo
df_temp.to_csv('clima_julho_tratado.csv', index=False)
print("üìÅ Arquivo tratado salvo como 'clima_julho_tratado.csv'")
```
