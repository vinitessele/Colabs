```python
from transformers import AutoProcessor, MedGEMMAForConditionalGeneration
from PIL import Image
import torch

# Carregamento do modelo
model = MedGEMMAForConditionalGeneration.from_pretrained("google/medgemma-4b-it", device_map="auto", torch_dtype=torch.float16)
processor = AutoProcessor.from_pretrained("google/medgemma-4b-it")

# Imagem de entrada
image = Image.open("mamografia_exemplo.png").convert("RGB")

# Prompt (pergunta)
prompt = "O que essa mamografia indica?"

# Processamento
inputs = processor(prompt, images=image, return_tensors="pt").to(model.device)

# Geração
generated_ids = model.generate(**inputs, max_new_tokens=256)
output = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print(output)
```
