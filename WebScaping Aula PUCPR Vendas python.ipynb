```python
# Instala a biblioteca BeautifulSoup e o requests
!pip install beautifulsoup4 requests

import requests
from bs4 import BeautifulSoup
```

```python
# URL do site que queremos fazer scraping
url = 'https://g1.globo.com/'

# Faz a requisição HTTP para a página
response = requests.get(url)

# Verifica se o acesso foi bem-sucedido (código 200)
if response.status_code == 200:
    print("Acesso bem-sucedido!")
else:
    print("Erro ao acessar a página:", response.status_code)
```

```python
# Usamos BeautifulSoup para interpretar o conteúdo da página
soup = BeautifulSoup(response.content, 'html.parser')

# Visualiza o início do HTML para análise
print(soup.prettify()[:1000])  # Mostra os primeiros 1000 caracteres
```

```python
# Vamos buscar todos os elementos que têm a classe das manchetes
# Inspecione o site para encontrar a classe correta (pode mudar!)
titulos = soup.find_all('a', class_='feed-post-link')  # classe das manchetes no G1

# Mostra os 5 primeiros títulos
for i, titulo in enumerate(titulos[:10]):
    print(f'{i+1}. {titulo.text.strip()}')
```

```python
import pandas as pd

# Criar lista com os textos das manchetes
lista_titulos = [titulo.text.strip() for titulo in titulos]

# Criar DataFrame com pandas
df = pd.DataFrame(lista_titulos, columns=["Título da Notícia"])

# Visualizar o DataFrame
df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://pt.wikipedia.org/wiki/Lista_de_países_por_população"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# A tabela que queremos é a primeira tabela com class "wikitable"
tabela = soup.find("table", {"class": "wikitable"})

# Parse da tabela para dataframe pandas
df = pd.read_html(str(tabela))[0]

df.head(30)
```
