```python
# 1. Instalar dependências
!apt install tesseract-ocr -y
!pip install pytesseract
!pip install --upgrade -q google-api-python-client google-auth-httplib2 google-auth-oauthlib
from google.colab import drive
import pytesseract
from PIL import Image
import os
import shutil

# 2. Montar o Google Drive
drive.mount('/content/drive')

# 3. Caminho para as imagens
pasta_drive = '/content/drive/MyDrive/imgCeonc/treinamento/images'

# 4. Função para processar OCR e renomear arquivos
def renomear_por_ocr(pasta):
    arquivos = [f for f in os.listdir(pasta) if f.lower().endswith('.png')]

    for nome in arquivos:
        caminho_img = os.path.join(pasta, nome)
        imagem = Image.open(caminho_img)

        # OCR
        texto = pytesseract.image_to_string(imagem, lang='por+eng')

        # Extrair primeira palavra em maiúsculo (ex: CCE, CCD, MLOE...)
        import re
        palavras = re.findall(r'\b[A-Z]{3,}\b', texto)

        if palavras:
            label = palavras[0]
            if not nome.upper().endswith(f'_{label}.PNG'):
                novo_nome = os.path.splitext(nome)[0] + f'_{label}.png'
                novo_caminho = os.path.join(pasta, novo_nome)

                os.rename(caminho_img, novo_caminho)
                print(f"✔️ Renomeado: {nome} → {novo_nome}")
            else:
                print(f"✅ Já contém: {nome}")
        else:
            print(f"❌ Nada detectado em: {nome}")

# 5. Rodar a função
renomear_por_ocr(pasta_drive)
```
