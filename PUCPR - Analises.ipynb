```python
# üìò An√°lise de Dados em 10 T√≥picos Did√°ticos
# Autor: Vinicius Tessele

# ================================
# üî¢ 1. Estat√≠sticas Descritivas
# ================================
import pandas as pd

print("\nüìä Estat√≠sticas Descritivas:")
df1 = pd.DataFrame({'idade': [22, 25, 30, 28, 24, 27]})
print(df1.describe())

```

```python
# ================================
# üìä 2. Distribui√ß√£o de Frequ√™ncia
# ================================
print("\nüìä Distribui√ß√£o de Frequ√™ncia:")
df2 = pd.DataFrame({'categoria': ['A', 'B', 'A', 'C', 'B', 'A']})
print(df2['categoria'].value_counts())
```

```python
# ================================
# ‚è≥ 3. Evolu√ß√£o Temporal
# ================================
import matplotlib.pyplot as plt

print("\nüìà Evolu√ß√£o Temporal:")
df3 = pd.DataFrame({
    'data': pd.date_range(start='2024-01-01', periods=5, freq='D'),
    'vendas': [100, 120, 130, 115, 140]
})
plt.plot(df3['data'], df3['vendas'], marker='o')
plt.title("üìà Vendas ao Longo do Tempo")
plt.xlabel("Data")
plt.ylabel("Vendas")
plt.grid()
plt.show()
```

```python
# ================================
# üîó 4. Correla√ß√£o entre Vari√°veis
# ================================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df4 = pd.DataFrame({
    'altura': [160, 170, 180, 175, 162, 165, 158, 172, 168, 174,
               163, 177, 169, 171, 166, 173, 161, 176, 164, 178,
               159, 167, 181, 170, 175, 162, 168, 174, 166, 172, 190],
    'peso': [60, 70, 80, 75, 61, 65, 55, 72, 68, 76,
             62, 78, 67, 71, 64, 74, 58, 77, 63, 79,
             56, 66, 82, 69, 75, 60, 67, 73, 65, 70, 85]
})

print(df4.corr())

sns.heatmap(df4.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, annot_kws={"size":16})
plt.title("Mapa de Calor da Correla√ß√£o")
plt.show()
```

```python
# ================================
# üîç 5. Agrupamentos e Compara√ß√µes
# ================================
print("\nüîç Comparando m√©dia salarial por sexo:")
df5 = pd.DataFrame({
    'sexo': ['M', 'F', 'F', 'M', 'F'],
    'salario': [3000, 3200, 2900, 3100, 3300]
})
print(df5.groupby('sexo')['salario'].mean())
```

```python
# ================================
# ‚ö†Ô∏è 6. An√°lise de Dados Ausentes
# ================================
print("\n‚ö†Ô∏è Dados Ausentes:")
df6 = pd.DataFrame({'nome': ['Ana', 'Bruno', None], 'idade': [25, None, 30]})
print(df6)
print("\nValores ausentes por coluna:")
print(df6.isnull().sum())
```

```python
# ================================
# üìä 7. Visualiza√ß√µes (Histograma)
# ================================
print("\nüìä Histograma de Notas:")
df7 = pd.DataFrame({'nota': [7, 8, 6, 9, 7, 8]})
sns.histplot(df7['nota'], bins=5, kde=True)
plt.title("Distribui√ß√£o de Notas")
plt.show()
```

```python
# ================================
# üìâ 8. Proje√ß√µes e Tend√™ncias
# ================================
from sklearn.linear_model import LinearRegression
import numpy as np

print("\nüìâ Proje√ß√£o com Regress√£o Linear:")
df8 = pd.DataFrame({'dia': [1, 2, 3, 4, 5], 'vendas': [100, 110, 115, 120, 130]})
X = np.array(df8['dia']).reshape(-1, 1)
y = df8['vendas']
modelo = LinearRegression().fit(X, y)
print(f"Previs√£o para o dia 6: {modelo.predict([[6]])[0]:.2f}")
```

```python

# ================================
# üß© 9. Segmenta√ß√£o e Ranking
# ================================
print("\nüß© Ranking de Clientes por Compras:")
df9 = pd.DataFrame({'cliente': ['A', 'B', 'C'], 'compras': [500, 1500, 800]})
df9['rank'] = df9['compras'].rank(ascending=False)
print(df9.sort_values('rank'))
```

```python
# ================================
# üö® 10. Detec√ß√£o de Outliers (Boxplot)
# ================================
print("\nüö® Detec√ß√£o de Outliers:")
df10 = pd.DataFrame({'nota': [5, 6, 7, 6, 100]})
sns.boxplot(x=df10['nota'])
plt.title("Boxplot de Notas (com outlier)")
plt.show()
```
